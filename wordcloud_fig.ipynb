```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data_dir = '/Users/alexbarnett/Documents/Community_detection/hippo_nets/'
```

```python
# these are the top terms after removing anatomical terms, redundant terms, disease names, and names of techniques (e.g. fMRI, restingstate)

context_terms = pd.read_excel(data_dir+'context_terms_358.xlsx')
DMN1_terms = pd.read_excel(data_dir+'DMN1_terms_358.xlsx')
DMN2_terms = pd.read_excel(data_dir+'DMN2_terms_358.xlsx')
DMN3_terms = pd.read_excel(data_dir+'DMN3_terms_358.xlsx')
```

```python
#MTN word cloud

context_wordcloud = WordCloud(background_color='white',color_func=lambda *args, **kwargs: ('#20943f'))

d = {}
for a, x in context_terms.values:
    d[a] = x
    
context_wordcloud.generate_from_frequencies(frequencies=d)
plt.figure()
plt.imshow(context_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig(data_dir+'context_cloud.png',dpi=300)
```
